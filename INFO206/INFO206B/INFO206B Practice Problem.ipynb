```python
def rec_sum(L):
    sum = 0
    if L == []:
        return 0
    else:
        if L[0]%2 ==0:
            sum = L[0]
        return sum + rec_sum(L[1:])
    
lil = [1,2,3,4,5]    
    
rec_sum(lil)
```

```python
def sumlistlist(lol):
    if lol==[]:
        return True
    else: 
        return sumlist(lol[0]) and sumlistlist(lol[1:])

def sumlist(l):
    sum = 0
    for i in range(len(l)):
        sum += l[i]
    return sum > 0

lol = [[1,2,3],[-7,6,-5],[4,3,5]]

sumlistlist(lol)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None# print all elements in a list
    def printList(self):
        cur = self;
        while cur != None:
            print(cur.data, end=' ') # print space-delimited
            cur = cur.next        
            print("") # print final carriage return# insert a node after the current node 
    
    def insertAfter(self, node): # insert a node at the end of the list 
        temp = self.next
        self.next = node
        node.next = temp
        
    def insertAtEnd(self, node)
        ptr = self.next
        while ptr.next != None:
             ptr = ptr.next
        ptr.insertAfter(node)
        
```

```python
def printSortedDictionary( D ):
    DS = sorted(D.items(), key=lambda x:x[1])
    for k in DS:
        print(k)# strip word of punctuation and convert to all lower-case
def stripWord( w ):
    w = w.replace( ".", "" )
    w = w.replace( ",", "" )
    w = w.replace( ";", "" )
    w = w.replace( ":", "" )
    w = w.replace( "'", "" )
    w = w.replace( "&", "" )
    w = w.replace( "\n", "" )
    w = w.lower()
    return( w )

import os

myfile = open('DOI.txt', 'r')
doi = {}

for line in myfile:
    for word in line.split():
        word = stripWord(word)
        if len(word) > 0:
            if word not in doi:
                doi[word] = 1
            else:
                doi[word] += 1
printSortedDictionary(doi)
```
