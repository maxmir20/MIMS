```python
import numpy as np
from numpy.linalg import inv, eig, norm
import matplotlib.pyplot as plt
import os

#real world + image coordinates
X_i =[100, 500,500,100]
Y_i = [100, 100, 300, 300]
u_i = [129.6, 169.5, 183.7, 142.9]
v_i = [216.5, 218.9, 86.4, 91.1]

plt.plot(u_i,v_i,'bo') # marks edges of license plate
img = plt.imread('license.jpg') #loads license plate
size = img.shape
plt.imshow(img)
plt.axis('off')
plt.show()

constraints = [] #our linear constraints matrix
for j in range(len(X_i)):
    constraint1 = [0 ,0 ,0 ,-X_i[j] ,-Y_i[j], -1, v_i[j]*X_i[j] ,v_i[j]*Y_i[j] ,v_i[j]]
    constraints.append(constraint1)
    
    constraint2 = [X_i[j] ,Y_i[j] ,1 ,0 ,0, 0, -u_i[j]*X_i[j] ,-u_i[j]*Y_i[j] , -u_i[j] ] 
    constraints.append(constraint2)
    
    
A = np.matrix(constraints) #creates a matrix out of our linear constraints

eigA = np.transpose(A)@A #uses X-transpose * X (in this case A-transpose * A ) to find the smallest eigenvector that will explain U

values, vectors = eig(eigA) #gives us the eigenvalues and eigenvectors 

h = vectors[:,np.argmin(values)] #pass the smallest eigenvector into h

H = np.reshape(h, (3, 3)) #reshapes h into 3x3 homography


import cv2 # you will need to install the opencv library: pip3 install opencv-python
im = plt.imread('license.jpg')
im_warp = cv2.warpPerspective(im, inv(H), (2*im.shape[1],im.shape[0]))
plt.imshow(im_warp) 
plt.gca().invert_yaxis()
plt.axis('off')
plt.show()

```
