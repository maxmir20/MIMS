```python
# %load quicksort.py
# Performs the quicksort algorithm.

# Partition the sublist the_list[p ... r] so that the pivot
# (originally in the_list[r]) moves to the_list[q],
# all items in the_list[p ... q-1] are less than or equal to the pivot,
# and all items in the_list[q+1 ... r] are greater than the pivot.
# Return the index q where the pivot ends up.
def partition(the_list, p, r, compare_func):
    pivot = the_list[r]
    
    # Set up the indices i and j so that
    #    the_list[p ... i] contains items <= pivot,
    #    the_list[i+1 ... j-1] contains items > pivot, and
    #    the_list[j ... r-1] contains items not yet compared with the pivot.
    i = p - 1
    j = p
    while j < r:
        if compare_func(the_list[j], pivot):
            # Move this item into the section known to be <= pivot.
            i += 1
            (the_list[i], the_list[j]) = (the_list[j], the_list[i])
        j += 1
    # Get the pivot into the correct position.
    (the_list[i+1], the_list[r]) = (the_list[r], the_list[i+1])
    #print(the_list[0], the_list[-1])
    return i+1

# Sort the sublist the_list[p ... r] using the quicksort algorithm.
def quicksort(the_list, p, r, compare_func):
    if p < r:   # nothing to do if the sublist has fewer than 2 items
        q = partition(the_list, p, r, compare_func) # divide
        quicksort(the_list, p, q-1, compare_func)   # conquer smaller items
        quicksort(the_list, q+1, r, compare_func)   # conquer larger items

# Sort the_list by running quicksort on it.        
def sort(the_list, compare_func):
    quicksort(the_list, 0, len(the_list)-1, compare_func)

# Return True if city1 has the same or higher population than city2.
def compare_population(city1, city2):
    #print(city1, city2)
    return city1.get_population() >= city2.get_population()

# Return True if city1 comes at or before city2 alphabetically.
def compare_name(city1, city2):
    return city1.get_name().lower() <= city2.get_name().lower()

# Return True if city1's latitude is less than or equal to city2's latitude.
def compare_latitude(city1, city2):
    return city1.get_latitude() <= city2.get_latitude()
```

```python
import os

class City:
    def __init__ (self, cc, name, region, pop, lat, long):
        #assigned using types to make sure there is no error in calculation
        self.cc = str(cc)
        self.name = str(name)
        self.region = str(region)
        self.pop = int(pop)
        self.lat = float(lat)
        self.long = float(long)
        
    def __str__(self):
         return (self.name + ", " + str(self.pop) + ", " + str(self.lat) + ", " + str(self.long))

    def get_name(self): #created for sort function
        return self.name
    
    def get_population(self): #created for sort function
        return self.pop
    
    def get_latitude(self): #created for sort function
        return self.lat
    
    def get_longitude(self): #created for sort function
        return self.long

citylist = [] #made to hold City references 
file = open('world_cities.txt', 'r') #opens the file
for i in file: #iterates through all lines in the file
    city = i.split(',') #splits each line into a list so we can reference them in creating the clas
    citylist.append(City(city[0], city[1], city[2], city[3], city[4], city[5])) #creates object of class City based on list 


file1 = open('cities_alpha.txt', 'w') #contains the list of cities sorted alphabetically.
sort(citylist, compare_name) #sorts the list 
for i in citylist: 
    file1.write(i.__str__() + '\n') #added \n so it will move to next line

file2 = open('cities_population.txt', 'w') #contains the list of cities sorted by population, from most to least populous.
sort(citylist, compare_population)
for i in citylist:
    file2.write(i.__str__() + '\n')#added \n so it will move to next line
    
file3 = open('cities_latitude.txt', 'w')#contains the list of cities sorted by latitude, from south to north.  
sort(citylist, compare_latitude)
for i in citylist:
    file3.write(i.__str__() + '\n')#added \n so it will move to next line
```

```python
# %load longlat_to_pixel.py
# Convert a latitude and longitude to pixel locations.
# WIDTH & HEIGHT should be defined to be the size of the image (in pixels)
def longlat_to_pixel(latitude, longitude):
    center_x = int(round((WIDTH / 2) + (WIDTH / 2) * (longitude / 180)))
    center_y = HEIGHT - int(round((HEIGHT / 2) + (HEIGHT / 2) * (latitude / 90)))
    return (center_x, center_y)
```

```python
import time
from IPython import display
import matplotlib.pyplot as plt

WIDTH = 720 # image (world.png) width
HEIGHT = 360 # image height
DPI = 60 # display image at this dots-per-inch resolution

pop_file = open('cities_population.txt', 'r') #opens the file for cities_population
pop_list = [] #empty list that will store the lines of pop_file as nested lists
for i in pop_file:
    i = i.strip() #removes \n from the end of each cities_population line
    sublist = i.split(',') #splits the line (which is a string) into a list based on each comma
    pop_list.append(sublist) #adds the sublist list as an element of pop_list

img = plt.imread("world.jpg")
plt_list = [] # used for holding all the previous plot points
for i in range(30): #will iterate 30 times to indicate the top 30
    plt.figure(figsize=(WIDTH/DPI,HEIGHT/DPI))
    plt.imshow(img) # Display the image
    plt.axis('off') 
    x,y = longlat_to_pixel(float(pop_list[i][2]), float(pop_list[i][3])) #uses the 3rd and 4th element of pop_list[i] to access the respective latitude and longitude
    if len(plt_list) != 0: #makes sure the list isn't empty before running through it
        for j in plt_list:
            plt.plot(j[0], j[1], 'bo') # plot one blue data point for each value stored in the list
    plt.plot(x, y, 'ro') # plot one red data point for the current list
    plt.text(x+5, y-8, str(pop_list[i][0]) ) #adds text 
    plt_list.append([x,y]) #adds to list of previous plot points so we can plot it in the next iteration
    time.sleep(.5)
    display.clear_output(wait=True)
    plt.show()

```
