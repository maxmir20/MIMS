```python
#!/usr/bin/env python3
# HW04_ch08_ex04

# The following functions are all intended to check whether a string contains
# any lowercase letters, but at least some of them are wrong. For each
# function, describe (is the docstring) what the function actually does.
# You can assume that the parameter is a string.

# Do not merely paste the output as a counterexample into the documentation
# string, explain what is wrong.


###############################################################################
# Body


def any_lowercase1(s):
    """Explain what is wrong, if anything, here.
    Does not track if there are lowercase letters in the string after the first letter. Once it is evaluated once, then it will
    only return one boolean value, even if the rest of the string is lowercase
    """ 
    for c in s:
        if c.islower():
            return True
        else:
            return False


def any_lowercase2(s):
    """Explain what is wrong, if anything, here.
    this only tests if the character 'c' is lowercase. It is not looking at the string at all
    """
    for c in s:
        if 'c'.islower():
            return 'True'
        else:
            return 'False'


def any_lowercase3(s):
    """Explain what is wrong, if anything, here.
    this updates the flag for every letter in the string, so the last update is the only one that is important. 
    E.g. if the last letter is uppercase, then it will return false, even if the first half of the string is lowercase
    """
    for c in s:
        flag = c.islower()
    return flag


def any_lowercase4(s):
    """Explain what is wrong, if anything, here.
    this one works because it will result in true if there is ever a lowercase letter in the string.
    When flag is deciding between false OR true, it will always go with true. If it goes with true then flag will become true.
    Any future evaluations will then be either flag = true or true, which will result in true, or flag = true or false, which
    will result in true. 
    """
    flag = False
    for c in s:
        flag = flag or c.islower()
    return flag


def any_lowercase5(s):
    """Explain what is wrong, if anything, here.
    This one will return false if it ever detects an uppercase number, even if the rest of the string is lowercase.
    """
    for c in s:
        if not c.islower():
            return False
    return True


###############################################################################
def main():

    # Remove print("Hello World!") and for each function above that is wrong,
    # call that function with a string for which the function returns
    # incorrectly.
    # ex.: any_lowercase_("thisstringmessesupthefunction")
    print("Hello World!")
    
    print(any_lowercase1('Underground'))
    print(any_lowercase2('UNDERGROUND'))
    print(any_lowercase3('undergrounD'))
    print(any_lowercase4('underGround'))
    print(any_lowercase5("underGround"))

if __name__ == '__main__':
    main()
```
