```python
#!/usr/bin/env python3
# HW05_ex00_TextAdventure.py
##############################################################################
# Imports
import sys

# Body

#yeah I know I shouldn't use global variables, but I didn't how to deal with
#classes yet and it seemed like the quick and dirty way to deal with something I could
#reference in multiple functions
inventory = []
        

def infinite_stairway_room(count=0):
    print("You are in a hallway. On one end is a staircase leading up towards light.")
    print('On the other end are some stairs leading back down')
    if count >= 3:
        print('you see an imp perched on the staircase')  
    loop = True
    while loop:
        next = input("> ")

        # infinite stairs option
        if count >=3:
            if next in "talk imp":
                print('\n the imps speaks:')
                print('"hehehe welcome to the endless staircase!"')
                print('"you`re trapped mortal, but I can help you escape if you answer my riddle"')
                answer = input('"The more you take away, the more I become. What am I?"')

                if answer in "a hole":
                    print('\n the imp vanishes in a puff of smoke') 
                    print('The room begins to blur and curve before melting away')
                    print('you`re back in the corridor\n')
                    hallway_two()
                else:
                      print('\n the imp snickers')
                      print('"hehehe not quite, see you on another floor"')
                      print('the imp bursts into flames and vanishes')
            if next in "insult imp":
                dead("The imp takes umbrage at your insult and shoots a fireball at your head. You are burned to a crisp")
            infinite_stairway_room(count=2)
        if next == "go up":
            print('You take the stairs, and emerge in a hallway')
            if (count > 0):
                print("but you're not happy about it")
            infinite_stairway_room(count + 1)
        elif next in "go back":
            print('You go back down the stairs, and emerge in a hallway')
            if (count > 0):
                print("but you're not happy about it")
            infinite_stairway_room(count + 1)              
        else: 
            print('sorry I didn`t quite get that. Please input another option')
    


def gold_room():
    print("passing through stalactites gently dripping water, you eventually are wading up to your waist")
    print('you emerge in a room filled with gold')
    print('How much do you take?')
    loop = True
    while loop:
        next = input("> ")
        try:
            how_much = int(next)
            loop = False
        except:
            print("that's not a number, try again.")
        

    if how_much < 50:
        print("Congratulations you win, enjoy your spoils!")
        exit()
    else:
        dead("You drown for taking too much gold. Nice job, genius")

def bat_room():
    print('you descend into a dark cave full of stalactites')
    print('there is a giant bat flapping menacingly before you')
    print('\n"greetings adventurer!" it shrieks')
    print('"so nice of you to drop in"')
    print('"shame you won`t be around for much longer!"')
    armor_damage = 0
    bat_damage = 0
    if 'armor' in inventory:
        while armor_damage < 6 and bat_damage < 2:
            print('\nthe bat swoops at you')
            print('the bat glances off your armor and leaves a dent in it')
            armor_damage += 1
            print('you can only survive for ' +str(6-armor_damage) + ' more hits before the armor breaks')
            print('beside the bat you see a stalactite that has broken off')
            next = input('> ')
            if next == 'inspect stalactite':
                print('it is a long stalactite with a very pointy end on it')
            elif next == 'take stalactite':
                print('you grab the stalactite with both hands and wield it')
                inventory.append('stalactite')
            elif next == 'stab bat':
                if 'stalactite' in inventory:
                    if bat_damage < 2:
                        print('you stab the bat and it reels in pain')
                        bat_damage += 1
                    elif bat_damage == 2:
                        print('the bat shrieks and collapses before you')
                        print('you step over its corpse and proceed to a glittery light deeper in the cave')
                        gold_room()
                else:
                    print('you don`t have anything to stab it with')
            else:
                print('not a valid action. Better move quickly before you get done in!')
        dead('the armor falls off from repeated assaults and you are ravaged by the bat. Better luck next time!')
    else:
        dead('the bat swoops at you and claws your head off')

def armor_room():
    if 'armor' not in inventory:
        print('in the room is a shiny suit of armor')
        print('it is locked in a cage in front of six levers')
        print('there is a sign next to the cage')
        print('behind you is the hallway door')
        loop = True
        unlock = False
        while loop:
            next = input("> ")

            if next in "read sign":
                print('you bend over to read the sign:')
                print('"as days become short"')
                print('"and nights become long"')
                print('"the harvest moon rises"')

            elif next == "inspect levers":
                print('the levers are organized in two rows of three')
                print('the top left lever (marked "1") is colored white')
                print('the top middle lever (marked "2") is shaped like an orb')
                print('the top right lever (marked "3") is colored black')
                print('the bottom left lever (marked "4") is colored orange')
                print('the bottom middle lever (marked "5") is shaped like a snake')
                print('the bottom right lever (marked "6") is shaped like a dwarf head')
            elif next == "pull 1" or next == "pull lever 1":
                print('you hear a thunk of gears turning')
                next = input("> ")
                if next == "pull 6" or next == "pull lever 6":
                    print('you hear a thunk of a second set of gears turning')
                    next = input("> ")
                    if next == "pull 3" or next == "pull lever 3":
                        print('you hear a thunk of a third set of gears turning')
                        next = input("> ")
                        if next == "pull 5" or next == "pull lever 5":
                            print('you hear a thunk of a fourth set of gears turning')
                            next = input("> ")
                            if next == "pull 4" or next == "pull lever 4":
                                print('you hear a thunk of a fifth set of gears turning')
                                next = input("> ")
                                if next == "pull 2" or next == "pull lever 2":
                                    print('you hear a thunk of a final set of gears turning')
                                    print('the cage clanks open and you don the set of armor')
                                    print('now you feel ready for anything')
                                    print('finding nothing else to do, you return to the corridor\n')
                                    inventory.append('armor')
                                    hallway_two()

                print('you hear a grinding of gears as the levers reset')
            elif next == "pull 2" or next == "pull 3" or next == "pull 4" or next == "pull 6" or next == "pull 5":
                print('you hear a grinding of gears as the levers reset')

            elif next == "hallway door":
                hallway_two()
            else:
                print('sorry I didn`t quite get that, please try again')
    else:
        print('it`s the room where you got the armor')
        print('there`s an empty cage and nothing else to do here')
        print('the hallway door is behind you')
        next = input('> ')
        while True:
            if next == 'hallway door':
                hallway_two()
            else:
                print('sorry I didn`t get that, please try something else')
      

        

def hallway_two():
    print('You emerge into a corridor that ends in two forks')
    print('the right fork is a staircase leading up, you can see a light at the end of it')
    print('the left fork leads downwards into darkness')
    print('there is a door on the the right wall of the corridor just before the fork')
    print('where do you go?')

    while True:
        next = input("> ")

        if next in "right fork":
            infinite_stairway_room()
        elif next in "left fork":
            bat_room()
        elif next in "right door":
            armor_room()
        else:
            print('sorry I didn`t quite catch that')
            
    
def bear_room():
    print("There is a bear here.")
    print("The bear has a bunch of honey.")    
    print("The bear is in front of a door.")
    print("How are you going to move the bear?")
    bear_moved = False

    while True:
        next = input("> ")

        if next == "take honey":
            dead("The bear looks at you then laughs at you.")
        elif next == "taunt bear" and not bear_moved:
            print("The bear has moved from the door. You can go through it now.")
            bear_moved = True
        elif next == "taunt bear" and bear_moved:
            dead("The bear gets angry and chases you.")
        elif next == "open door" and bear_moved:
            hallway_two()
            #gold_room()
        else:
            print("I got no idea what that means.")


def cthulhu_room():
    print("Here you see the great evil Cthulhu.")
    print("He, it, whatever stares at you and you freeze.")
    print("Do you flee for your life or try to make friends?")

    next = input("> ")

    if "flee" in next:
        start()
    elif "friends" in next:
        dead("Well that was tasty!")
    else:
        cthulhu_room()


def dead(why):
    print("{}\n Game Over!".format(why))
    try:
        choice = input('New Game?...y/n ')
        if choice == 'y':
            start()
        else:
            exit()
    except:
        exit()

############################################################################
def start():
    # START the TextAdventure game
    user = input('greetings adventurer!, what is your name: ')
    print('welcome ' + user + ' your journey awaits')
    print("You awaken in a dark room.")
    print("There is a door to your right and left.")
    print("Which one do you take?")

    next = input("> ")
    if next == "left":
        bear_room()
    elif next == "right":
        cthulhu_room()
    else:
        dead("You stumble around the room until you starve.")


if __name__ == '__main__':
    start()
```
