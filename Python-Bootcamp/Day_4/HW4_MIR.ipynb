```python
#!/usr/bin/env python3

# (1)
# Write a function called has_no_e that returns True if the given word doesn't
# have the letter "e" in it.
#   - write has_no_e
# (2)
# Modify your program from 9.1 to print only the words that have no "e" and
# compute the percentage of the words in the list that have no "e".
#   - print each approved word on new line, followed at the end by the %
#   - name your function print_no_e
##############################################################################
# Imports

# Body
def no_e(a):
    e_count = 0
    for i in a:
        if i == 'e':
            e_count += 1
    if e_count == 0:
        return True
    else:
        return False
    
def print_no_e(a):
    a_string = a.split()
    e_list = []
    e_count = 0
    e_total = len(a_string)
    for i in a_string:
        
        if no_e(i):
            e_list.append(i) 
            e_count += 1

    for i in e_list:
        print(i)
    print(str((e_count/e_total) * 100) + '% do not contain e')

##############################################################################
def main():
    pass  # Call your function(s) here.
    print_no_e('the quick brown fox jumped over the lazy dog')
    print_no_e('I wish I had a furry cat')
    print_no_e('eenie meanie miney moe')



if __name__ == '__main__':
    main()
```

In this case I found it was easier to make a modified version of the forbidden_prompt function called forbidden_prompt_alt that would return a count of excluded words instead. I also made it so it could accept a string argument, so forbidden letters could be passed into it. 

2. Finding smallest combination: this was difficult for a number of reasons. Originally, I thought the best solution was to evaluate each letter to identify the top 5 letters that excluded the fewest amount of words. Once you had those letters, you can combine them into a string and run that through and see how many were excluded. This led to other problems, Which are identified below.
    
    a. Sorting: In order to determine the top 5 least excluded letters, you would need to first run each letter to return all the numbers, then sort from lowest to highest. If you don't tie it to the letters however, then you won't be able to identify it once you've sorted them. The solution in this case was to create a nested list that contains both the letter and the output, then sorting only by the output. 
    
    b. Duplicates: While I originally thought that that solution would be adequate for the problem, it was pointed out to me by my classmates that this wouldn't account for instances in which letters could overlap each other in terms of words excluded. Even by taking the combination smallest letters, it was possible that they would each be excluding separate words, wheres another combination would have more overlap in the words they excluded, potentially lowering the total number of excluded words. In order to test this, I worked with my classmates to combine two forms of code. The first one, noted above, would find the top 5 least excluded letters, then generate the total number from that combination. Using that number as a baseline, we would then compare that against the number of excluded words for each letter. If any letters excluded more words than the combination, it was disqualified. Using this new list of letters than fell under the baseline, I used an itertools.combinations function to create a list of all possible variations and compared each result against the baseline. If a combination was found that would be greater, then it would replace the existing baseline and form a new one.

```python
#!/usr/bin/env python3

# (1)
# Write a function named avoids that takes a word and a string of forbidden
# letters, and that returns True if the word doesn't use any of the forbidden
# letters.
#   - write avoids
# (2)
# Modify your program to prompt the user to enter a string of forbidden
# letters and then print the number of words that don't contain any of them.
#   - write forbidden_prompt and
#   - modify to create forbidden_param that accepts the string as an argument
# (3)
# Can you find a combination of 5 forbidden letters that excludes the smallest
# number of words?
#   - write a function that finds this combination of letters: find_five
#   - have that function print the letters and print the # of words excluded
##############################################################################
# Imports
import os
import string
import itertools
# Body

def avoids(a,f):
    """ return True if word NOT forbidden"""
    forbidden = list(f)
    no_forbid= True
    for i in a:
        if i in forbidden:
            no_forbid = False
    return no_forbid
    pass

def forbidden_prompt():
    """ print count of words NOT forbidden by input"""
    forbidden = input('please input a string of forbidden letters ')
    file = open('words.txt', 'r')
    cleared_words = []
    cleared_count = 0
    for i in file:
        if avoids(i, forbidden):
            cleared_words.append(i)
            cleared_count += 1
    print(cleared_count)
    pass


#used for the third exercise. Modified so it will take an argument
#and will return the list of excluded words instead of cleared words
def forbidden_prompt_alt(f):
    """ print count of words forbidden by input"""
    forbidden = f
    file = open('words.txt', 'r')
    cleared_words = []
    #count used for find_five function
    excluded_count = 0
    for i in file:
        if avoids(i, forbidden):
            cleared_words.append(i)
        else:
            excluded_count += 1
    return(excluded_count)
    pass


#used to sort the forbid_list list in find_five by the [0] index position
def sortFirst(val):
    return val[0]

def find_five():
    forbid_list = []
    for i in string.ascii_lowercase:
        # runs each letter through forbidden_prompt_alt, which returns number of excluded words
        # then adds a list of the # of excluded words and the letter used
        ascii_link = [forbidden_prompt_alt(i), i]
        #then adds that list to the forbid_list, which will be used for comparing letters
        forbid_list.append(ascii_link)
    #sorts the list by the [0] index position of the nested list in each index position of forbid_list
    forbid_list.sort(key = sortFirst)
    forbid_str = ''
    forbid_cnt = 0
    #takes the letters assigned to the 5 least excluded numbers and adds them to a string
    for i in range(5):
        forbid_str += (forbid_list[i][1])

    #runs the string of top 5 letters to create an upper bound to compare individual characters against
    upper_bound = forbidden_prompt_alt(forbid_str)
    #creates a second list to hold letters in forbid_list that exclude less than the combination string of the top 5 
    second_run = []
    for i in forbid_list:
        if i[0] < upper_bound:
            second_run.append(i[1])
    """ 
    Now that we have our list of letters and upper bound to compare against, we run a program to combine
    all variations of the letters to see if any of them generate a lower count of excluded words than our
    upper bound. If they do, then we replace the final count and string with the next lowest count and keep
    going until we've reached ever combination. After which we print out the final lowest number
    """
    final_string = forbid_str
    final_count = upper_bound
    restr_combo = itertools.combinations(second_run,5)
    for i in restr_combo:
        comp_string = ''.join(i)
        comp_count = forbidden_prompt_alt(comp_string)
        #used to indicate when a string has been replaced
        if final_count > comp_count:
            print('new string is smaller')
            print('old string: ' + final_string + ' ; old count: ' + str(final_count))
            print('new string: ' + comp_string + ' ; new string: ' + str(comp_count))
            final_count = comp_count
            final_string = comp_string
    #prints final tally once every combination has been reached
    print(' the final string that yields the smallest combo is ' + final_string)
    print(' this string excludes ' + str(final_count) + ' words')

    
##############################################################################
def main():
    find_five()
    # Your final submission should only call five_five


if __name__ == '__main__':
    main()
```
